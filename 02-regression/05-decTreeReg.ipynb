```r
# Importing Dataset
dir = '../Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 8 - Decision Tree Regression/R/Position_Salaries.csv'
dataset = read.csv(dir)
```

```r
dataset
```

```r
dataset = dataset[2:3]
```

```r
#install.packages('rpart')
library('rpart')
```

```r
regressor = rpart(formula = Salary ~ .,
                  data = dataset,
                  control = rpart.control(minsplit = 1))
```

```r
y_pred = predict(regressor, data.frame(Level = 6.5))
y_pred
```

```r
library('ggplot2')
x_grid =  seq(min(dataset$Level), max(dataset$Level), 0.1)
```

```r
ggplot() +
    geom_point(aes(x=dataset$Level, y=dataset$Salary),color='red') +
    geom_line(aes(x=dataset$Level, y=predict(regressor, newdata=dataset)), color='orange') +
    ggtitle('Truth or Bluff (Decision Tree)') +
    xlab('Level')
```
