```r
# Importing Dataset
dir = '../Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 9 - Random Forest Regression/R/Position_Salaries.csv'
dataset = read.csv(dir)
```

```r
dataset
```

```r
dataset = dataset[2:3]
```

```r
#install.packages('randomForest')
library('randomForest')
set.seed(1234)
```

```r
rfNews()
```

```r
regressor = randomForest(x = dataset[1],
                         y = dataset$Salary,
                         ntree = 10)
```

```r
y_pred = predict(regressor, data.frame(Level = 6.5))
y_pred
```

```r
library('ggplot2')
x_grid =  seq(min(dataset$Level), max(dataset$Level), 0.001)
```

```r
ggplot() +
    geom_point(aes(x=dataset$Level, y=dataset$Salary),color='red') +
    geom_line(aes(x=dataset$Level, y=predict(regressor, newdata=dataset)), color='orange') +
    ggtitle('Truth or Bluff (Random Forest)') +
    xlab('Level')
```
