```r
# Importing Dataset
dir = '../Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 5 - Multiple Linear Regression/R/50_Startups.csv'
dataset = read.csv(dir)
```

```r
dataset
```

```r
dataset$State
```

```r
dataset$State = factor(dataset$State,
                        levels = c('California', 'Florida', 'New York'),
                        labels = c( 1, 2, 3))
```

```r
dataset
```

```r
#SPLITTING THE DATA
#install.packages("caTools")
```

```r
library('caTools')
set.seed(123)
```

```r
split = sample.split(dataset$Profit, SplitRatio=0.8 )
training_set = subset(dataset, split == TRUE)
test_set = subset(dataset, split==FALSE)
```

```r
training_set
```

```r
test_set
```

```r
regressor = lm(formula = Profit ~ .,
              data = training_set
              )
```

```r
summary(regressor)
```

```r
y_pred = predict(regressor, newdata = test_set)
```

```r
y_pred
```

```r
#BACKWARD ELIMINATION IS ELIMINATING NON STATISTICALLY SIGNIFICANT VARIABLE
```

```r
regressor = lm(formula = Profit ~ + R.D.Spend + Marketing.Spend  ,
              data = dataset
              )
summary(regressor)

#eliminate State
```

```r
y_pred = predict(regressor, newdata = test_set)
y_pred
```
