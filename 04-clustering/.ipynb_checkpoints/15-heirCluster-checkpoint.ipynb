```r
# Importing Dataset
dir = '../../pynb/04-clustering/Mall_Customers.csv'
dataset <- read.csv(dir)
```

```r
X <- dataset[4:5]
```

```r
dendogram = hclust(dist(X, method = 'eucli'))
```

```r
wcss <- vector()
```

```r
for (i in 1:10) wcss[i] <- sum(kmeans(X,i)$withinss)
plot(1:10, wcss, type = 'b', main = paste("Main Cluster"), xlab = 'Number', ylab='wcss')
```

```r
set.seed(29)
```

```r
kmeans <- kmeans(X, 5, iter.max =300, nstart=10)
```

```r
library(cluster)
```

```r
clusplot(X,
         kmeans$cluster,
         lines = 0,
         shade = TRUE,
         color = TRUE,
         #labels = 2,
         plotchar = FALSE,
         span = TRUE,
         main = paste('Cluster of Clients'),
         ylab = 'Spending Score',
         xlab = 'Annual Income')
          
```
