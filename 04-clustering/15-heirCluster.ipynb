```r
# Importing Dataset
dir = '../../pynb/04-clustering/Mall_Customers.csv'
dataset <- read.csv(dir)
```

```r
X <- dataset[4:5]
```

```r
dendogram = hclust(dist(X, method = 'euclidean'), method = 'ward.D')
```

```r
plot (dendogram,
      main = paste('Dendogram'),
      xlab = 'Customers',
      ylab = 'Euclidan Distance')
```

```r
hc =  hclust(dist(X, method = 'euclidean'), method = 'ward.D')
```

```r
y_hc = cutree(hc, 5)
```

```r
y_hc
```

```r
library(cluster)
```

```r
clusplot(X,
         y_hc,
         lines = 0,
         shade = TRUE,
         color = TRUE,
         #labels = 2,
         plotchar = FALSE,
         span = TRUE,
         main = paste('Cluster of Clients'),
         ylab = 'Spending Score',
         xlab = 'Annual Income')
          
```
